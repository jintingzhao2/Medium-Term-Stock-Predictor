```python
pip install yfinance matplotlib import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd


def get_stock_price_and_plot(user_input):
    """
    Fetches stock data from Yahoo Finance based on company name or ticker,
    prints current price, and plots 1-year closing price chart.
    """

    # Try searching Yahoo Finance if user enters company name
    search = yf.Search(user_input)
    if search.quotes:
        ticker_symbol = search.quotes[0]["symbol"]
    else:
        print("Stock not found.")
        return

    print(f"Using ticker: {ticker_symbol}")

    # Download 1 year of data
    stock = yf.Ticker(ticker_symbol)
    hist = stock.history(period="1y")

    if hist.empty:
        print("No historical data available.")
        return

    # Get current price
    current_price = hist["Close"].iloc[-1]
    print(f"Current Price: ${current_price:.2f}")

    # Plot line chart
    plt.figure()
    plt.plot(hist.index, hist["Close"])
    plt.title(f"{ticker_symbol} - Last 1 Year Closing Prices")
    plt.xlabel("Date")
    plt.ylabel("Closing Price ($)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

```

```python
import streamlit as st
import yfinance as yf
import matplotlib.pyplot as plt
```

```python

st.title("📈 Stock Price Viewer")

user_input = st.text_input("Enter Company Name or Ticker:")

if user_input:
    search = yf.Search(user_input)

    if search.quotes:
        ticker_symbol = search.quotes[0]["symbol"]
        st.write(f"Using ticker: **{ticker_symbol}**")

        stock = yf.Ticker(ticker_symbol)
        hist = stock.history(period="1y")

        if not hist.empty:
            current_price = hist["Close"].iloc[-1]
            st.write(f"Current Price: **${current_price:.2f}**")

            fig, ax = plt.subplots()
            ax.plot(hist.index, hist["Close"])
            ax.set_title(f"{ticker_symbol} - Last 1 Year Closing Prices")
            ax.set_xlabel("Date")
            ax.set_ylabel("Closing Price ($)")
            st.pyplot(fig)
        else:
            st.error("No historical data found.")
    else:
        st.error("Stock not found.")
```
